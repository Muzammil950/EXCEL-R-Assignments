```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("sales_data_with_discounts (1).csv")
```

```python
df.head()
```

```python
# 1. Descriptive Statistics
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
# 2. Histograms
df.hist(figsize=(10,8))
plt.show()
```

```python
# 3. Boxplots
for col in df.select_dtypes(include='number').columns:
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
# 4. Bar chart for categorical columns
for col in df.select_dtypes(include='object').columns:
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

## Conclusion
### The dataset shows variation in sales and discounts.
### Some columns have outliers as seen in boxplots.
### The distribution of values is slightly skewed.
### Categorical data shows uneven distribution among categories.
