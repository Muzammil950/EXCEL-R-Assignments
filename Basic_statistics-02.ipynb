```python
import pandas as pd
```

```python
# 1. Load datasets
patients = pd.read_csv("Patient_Data.csv")
billing = pd.read_csv("Billing_Data.csv")
```

```python
patients.head()
```

```python
billing.head()
```

```python
# Show summary
print("Patient Info:\n")
print(patients.info())
```

```python
# 2. Select relevant columns
patients = patients[['PatientID', 'Department', 'Doctor', 'BillAmount']]
```

```python
patients.head()
```

```python
# 3. Drop unnecessary columns (safe)
patients = patients.drop(columns=['ReceptionistID', 'CheckInTime'], errors='ignore')
```

```python
patients.head()
```

```python
# 4. Groupby - total bill per department
dept_bill = patients.groupby('Department')['BillAmount'].sum()
print("\nTotal Bill per Department:\n", dept_bill)
```

```python
# 5. Remove duplicates
patients = patients.drop_duplicates(subset='PatientID')
```

```python
# 6. Fill missing BillAmount with mean
patients['BillAmount'].fillna(patients['BillAmount'].mean(), inplace=True)
```

```python
# 7. Merge datasets
merged_df = pd.merge(patients, billing, on='PatientID', how='inner')
```

```python
# 8. Add new patients (row-wise)
new_patients = pd.DataFrame({
    'PatientID': [999, 1000],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr X', 'Dr Y'],
    'BillAmount': [8000, 9000]
})
```

```python
final_df = pd.concat([merged_df, new_patients], axis=0, ignore_index=True)
```

```python
# 9. Add new columns (column-wise)
final_df['InsuranceCovered'] = True
final_df['FinalAmount'] = final_df['BillAmount'] * 0.9
```

```python
# Final Output
print("\nFinal Cleaned Dataset:\n")
print(final_df)
```

## conclusion
- The dataset was cleaned by removing duplicates and handling missing values.
- Data from multiple sources was merged using PatientID.
- Department-wise billing analysis was performed using groupby.
- Additional patient data and billing features were added for better analysis
