### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np
```

```python
# Create DataFrame with 10 rows
data = {
    "Name": ["baksh", "amir", "janisar", "izmamul", "arif", "atif", "maruf", "sharuk", "qasim", "adil"],
    "Age": [22, 25, 23, 24, 26, 28, 21, 27, 29, 22],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi", "Chennai", "Kolkata", "Delhi", "Mumbai", "Pune", "Delhi"],
    "Salary": [30000, 40000, np.nan, 35000, 45000, np.nan, 32000, 41000, 38000, 30000]
}
```

```python
df = pd.DataFrame(data)

print("Original DataFrame:\n", df)
```

```python
# 1. info() and describe()
print("\nInfo:")
print(df.info())
```

```python
print("\nDescribe:")
print(df.describe())
```

```python
# 2. Select 'Name' and 'City'
print("\nSelected Columns:")
print(df[["Name", "City"]])
```

```python
# 3. Drop 'City' column
df = df.drop("City", axis=1)
print("\nAfter Dropping City:\n", df)
```

```python
# 4. Fill missing values in Salary with mean
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
print("\nAfter Filling Missing Salary:\n", df)
```

```python
# 5. Remove duplicate rows
df = df.drop_duplicates()
print("\nAfter Removing Duplicates:\n", df)
```
