# Recommendation System using Cosine Similarity

This notebook implements an Anime Recommendation System using cosine similarity.

```
# Import Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```
# Load Dataset
df = pd.read_csv("anime.csv")

# First 5 rows
df.head()
```

```
# Dataset Information
df.info()
```

```
# Statistical Summary
df.describe(include='all')
```

```
# Check Missing Values
df.isnull().sum()
```

```
# Handle Missing Values
df.dropna(inplace=True)
```

```
# Check Duplicate Values
df.duplicated().sum()
```

```
# Remove Duplicate Values
df.drop_duplicates(inplace=True)
```

```
# Distribution of Ratings
plt.figure(figsize=(8,5))

sns.histplot(df['rating'], bins=20, kde=True)

plt.title("Distribution of Anime Ratings")

plt.show()
```

```
# Top Genres
plt.figure(figsize=(10,6))

df['genre'].str.split(',').explode().value_counts().head(10).plot(kind='bar')

plt.title("Top Genres")

plt.show()
```

```
# Feature Selection
features = df[['name', 'genre', 'rating', 'members']]

features.head()
```

```
# Normalize Numerical Features
scaler = MinMaxScaler()

features[['rating', 'members']] = scaler.fit_transform(
    features[['rating', 'members']]
)
```

```
# Convert Genre into Numerical Features
features['genre'] = features['genre'].astype(str)

cv = CountVectorizer(stop_words='english')

genre_matrix = cv.fit_transform(features['genre'])
```

```
# Compute Cosine Similarity
cosine_sim = cosine_similarity(genre_matrix)

print(cosine_sim.shape)
```

```
# Create Index Mapping
indices = pd.Series(
    features.index,
    index=features['name']
).drop_duplicates()
```

```
# Recommendation Function
def recommend_anime(title, cosine_sim=cosine_sim):

    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(
        sim_scores,
        key=lambda x: x[1],
        reverse=True
    )

    sim_scores = sim_scores[1:6]

    anime_indices = [i[0] for i in sim_scores]

    return df['name'].iloc[anime_indices]
```

```
# Example Recommendation
recommend_anime("Naruto")
```

```
# Threshold Example
title = "Naruto"

idx = indices[title]

scores = list(enumerate(cosine_sim[idx]))

threshold = 0.5

filtered = [x for x in scores if x[1] > threshold]

print("Recommendations above threshold:", len(filtered))
```

```
# Ratings vs Members Visualization
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df['rating'],
    y=df['members']
)

plt.title("Ratings vs Members")

plt.show()
```

```
# Another Recommendation Example
recommend_anime("Death Note")
```

## 1. Difference between User-Based and Item-Based Collaborative Filtering

### User-Based Collaborative Filtering
- Recommends items liked by similar users.

### Item-Based Collaborative Filtering
- Recommends items similar to already liked items.
